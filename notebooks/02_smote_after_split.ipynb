# **Applied Machine Learning for Cyber Security**

## **Notebook 2: Modeling with SMOTE Applied After Train-Test Split**

This notebook is part of the HELLO_FLOOD attack detection project. It focuses on improving classification performance by addressing class imbalance using **SMOTE (Synthetic Minority Over-sampling Technique)**, applied **after splitting the dataset into training and testing sets**.

The goal is to explore how applying SMOTE only on the training data impacts supervised model performance while ensuring a realistic, untouched test set. Additionally, a basic unsupervised anomaly detection technique is implemented for exploratory purposes.

---

### **Covered in this notebook:**
- Preprocessing and cleaning of the HELLO_FLOOD dataset
- Label encoding and feature scaling
- Stratified train-test split (70/30)
- **Application of SMOTE on training set only, followed by data shuffling**
- Implementation and evaluation of the following supervised models:
  - Logistic Regression
  - Random Forest Classifier
  - Multi-Layer Perceptron (Neural Network)
  - LSTM (Long Short-Term Memory) using PyTorch
  - GRU (Gated Recurrent Unit) using PyTorch
- **Unsupervised anomaly detection using Isolation Forest**
- Evaluation metrics: Accuracy, Precision, Recall, F1 Score
- Confusion matrix visualizations and random prediction samples
- Performance comparison using:
  - Tabular summary
  - Bar plot
  - F1 Score slope graph
---
This notebook serves as a key checkpoint to assess the effectiveness of SMOTE applied post split and compare its outcomes with the baseline approach (Notebook 1).

### **3. Model Building, Implementation and Evaluation**

#### **3.1 Imputing NaN Values**

```python
# Impute remaining NaNs with column means (recommended for numeric features)
X_scaled_df.fillna(X_scaled_df.mean(), inplace=True)

console.print("[bold green]NaN values filled using column mean.[/bold green]")
console.print(f"[bold blue]New shape after imputation:[/bold blue] {X_scaled_df.shape}")
```

#### **3.2 Data Splitting using Test-Train Split**

```python
# Use scaled features (X_scaled_df) and target labels (y)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled_df, y, test_size=0.3, random_state=42, stratify=y
)

console.print("[bold green]Train-test split completed.[/bold green]")
console.print(f"[bold blue]Training Set:[/bold blue] {X_train.shape[0]} samples",
              f"\n[bold blue]Testing Set :[/bold blue] {X_test.shape[0]} samples")
```

#### **3.3 Applying SMOTE to the Training Set (After Splitting)**

```python
# Apply SMOTE on training data
sm = SMOTE(random_state=42)
X_train_smote, y_train_smote = sm.fit_resample(X_train, y_train)

# Shuffle after SMOTE
X_train_smote, y_train_smote = shuffle(X_train_smote, y_train_smote, random_state=42)

console.print(f"[bold green]After SMOTE - Training set size:[/bold green] {X_train_smote.shape}")
console.print(y_train_smote.value_counts())

# Reset index for both to avoid mismatch
X_train_smote = X_train_smote.reset_index(drop=True)
y_train_smote = y_train_smote.reset_index(drop=True)
```

#### **3.4 Helper Function for Model Evaluation**

```python
# Helper to evaluate model
def evaluate_model(model, model_name, X_test, y_test, y_pred):
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    console.print(f"[bold]{model_name} Evaluation (SMOTE After Split)[/bold]",
                  f"\n[bold]Accuracy :[/bold] {acc:.4f}",
                  f"\n[bold]Precision:[/bold] {prec:.4f}",
                  f"\n[bold]Recall   :[/bold] {rec:.4f}",
                  f"\n[bold]F1 Score :[/bold] {f1:.4f}")

    # Confusion matrix
    plt.figure(figsize=(4, 3))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues")
    plt.title(f"{model_name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

    # First 10 predictions (random sample)
    sample_idx = np.random.choice(len(y_test), size=10, replace=False)
    comparison_df = pd.DataFrame({
        'Actual': y_test.values[sample_idx],
        'Predicted': y_pred[sample_idx]
    })
    console.print(f"[bold]Random 10 Predictions - {model_name}[/bold]")
    console.print(comparison_df)
```

##### **3.5.4 LSTM (Long-Short Term Memory)**

```python
# ============================================================================================================
# Step 1: Prepare Sequential Data from SMOTE-applied training set
# ============================================================================================================
def create_lstm_sequences(X, y, window_size=10):
    X_seq, y_seq = [], []
    for i in range(len(X) - window_size):
        X_seq.append(X[i:i+window_size].values)
        y_seq.append(y[i+window_size])
    return np.array(X_seq), np.array(y_seq)

# Use the concatenated and scaled training set
X_combined = pd.concat([X_train_smote, X_test], axis=0).reset_index(drop=True)
y_combined = pd.concat([y_train_smote, y_test], axis=0).reset_index(drop=True)

X_lstm, y_lstm = create_lstm_sequences(X_combined, y_combined, window_size=10)

X_tensor = torch.tensor(X_lstm, dtype=torch.float32)
y_tensor = torch.tensor(y_lstm, dtype=torch.float32)
```

```python
# ============================================================================================================
# Step 2: Train-Test Split for Sequences
# ============================================================================================================

# Split into 70% train / 30% test
train_size = int(0.7 * len(X_tensor))
test_size = len(X_tensor) - train_size

train_dataset, test_dataset = random_split(TensorDataset(X_tensor, y_tensor), [train_size, test_size])
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64)
```

```python
# ============================================================================================================
# Step 3: Define LSTM Model
# ============================================================================================================

class LSTMClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(LSTMClassifier, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        out, _ = self.lstm(x)
        out = out[:, -1, :]
        out = self.fc(out)
        return self.sigmoid(out)

input_size = X_combined.shape[1]
lstm_model = LSTMClassifier(input_size=input_size, hidden_size=64, num_layers=2)

criterion = nn.BCELoss()
optimizer = torch.optim.Adam(lstm_model.parameters(), lr=0.001)
```

```python
# ============================================================================================================
# Step 4: Train the LSTM Model
# ============================================================================================================

lstm_model.train()
epochs = 5
loss_history = []

for epoch in range(epochs):
    epoch_loss = 0
    for batch_x, batch_y in train_loader:
        optimizer.zero_grad()
        output = lstm_model(batch_x).squeeze()
        loss = criterion(output, batch_y)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    avg_loss = epoch_loss / len(train_loader)
    loss_history.append(avg_loss)
    console.print(f"[bold green]LSTM Epoch {epoch+1}/{epochs}, LSTM Loss: {avg_loss:.4f}[/bold green]")
```

```python
# ============================================================================================================
# Step 5: Plot Loss Curve
# ============================================================================================================

plt.figure(figsize=(10, 4))
plt.plot(range(1, epochs + 1), loss_history, marker='o', color='purple')
plt.title("LSTM Training Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ============================================================================================================
# Step 6: Evaluate the LSTM Model
# ============================================================================================================

lstm_model.eval()
all_preds, all_labels = [], []

with torch.no_grad():
    for batch_x, batch_y in test_loader:
        output = lstm_model(batch_x).squeeze()
        preds = (output > 0.5).float()
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(batch_y.cpu().numpy())

y_pred_lstm = pd.Series(np.array(all_preds).astype(int))
y_test_lstm = pd.Series(np.array(all_labels).astype(int))

# Use your existing evaluation helper
evaluate_model(model=None, model_name="LSTM (SMOTE After Split)", X_test=None, y_test=y_test_lstm, y_pred=y_pred_lstm)
```

##### **3.5.5 GRU (Gated Recurrent Units)**

```python
# ============================================================================================================
# Step 1: Prepare Sequential Data and apply Train-Test Split
# ============================================================================================================

# Reuse these from LSTM section
# X_lstm, y_lstm
# X_tensor, y_tensor
# train_dataset, test_dataset
# train_loader, test_loader
# So no need to duplicate the sequence creation logic — GRU will use the same train_loader and test_loader as LSTM.
```

```python
# ============================================================================================================
# Step 2: Define the GRU Model
# ============================================================================================================

class GRUClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(GRUClassifier, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        out, _ = self.gru(x)
        out = out[:, -1, :]
        out = self.fc(out)
        return self.sigmoid(out)

gru_model = GRUClassifier(input_size=input_size, hidden_size=64, num_layers=2)

criterion_gru = nn.BCELoss()
optimizer_gru = torch.optim.Adam(gru_model.parameters(), lr=0.001)
```

```python
# ============================================================================================================
# Step 3: Train GRU Model
# ============================================================================================================

gru_model.train()
gru_loss_history = []
epochs = 5

for epoch in range(epochs):
    epoch_loss = 0
    for batch_x, batch_y in train_loader:
        optimizer_gru.zero_grad()
        output = gru_model(batch_x).squeeze()
        loss = criterion_gru(output, batch_y)
        loss.backward()
        optimizer_gru.step()
        epoch_loss += loss.item()
    
    avg_loss = epoch_loss / len(train_loader)
    gru_loss_history.append(avg_loss)
    console.print(f"[bold green]GRU Epoch {epoch+1}/{epochs}, GRU Loss: {avg_loss:.4f}[/bold green]")
```

```python
# ============================================================================================================
# Step 4: Plot Training Loss
# ============================================================================================================

plt.figure(figsize=(10, 4))
plt.plot(range(1, epochs + 1), gru_loss_history, marker='o', color='teal')
plt.title("GRU Training Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ============================================================================================================
# Step 5: Evaluate GRU Model
# ============================================================================================================

gru_model.eval()
gru_preds, gru_labels = [], []

with torch.no_grad():
    for batch_x, batch_y in test_loader:
        output = gru_model(batch_x).squeeze()
        preds = (output > 0.5).float()
        gru_preds.extend(preds.cpu().numpy())
        gru_labels.extend(batch_y.cpu().numpy())

y_pred_gru = pd.Series(np.array(gru_preds).astype(int))
y_test_gru = pd.Series(np.array(gru_labels).astype(int))

evaluate_model(model=None, model_name="GRU (SMOTE After Split)", X_test=None, y_test=y_test_gru, y_pred=y_pred_gru)
```

##### **3.4.6 Isolation Forest**

```python
# ============================================================================================================
# Step 1: Filter benign samples for training
# ============================================================================================================

X_benign = X_scaled_df[y == 0]
X_full = X_scaled_df
y_true = y.reset_index(drop=True)
```
